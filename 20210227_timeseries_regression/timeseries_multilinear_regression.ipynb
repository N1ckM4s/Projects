```
import pandas as pd
import numpy as np

###PERCHE I MODELLI DI CLASSIFICAZIONE NON HANNO QUESTO PROBLEMA?
###TUTTI I MODELLI DI REGRESSIONE HANNO LO STESSO PROBLEMA?

#PERCHE LA REGRESSIONE FUNZIONI BISOGNA DARE AL MODELLO GLI INPUT INDIVIDUALMENTE: 
#NON [1, 2, 3, 4, 5], altrimenti non posso neanche inserire un input alla volta, ma solo un input di 5
#[[1], [2], [3], [4], [5]

#linear regression on timeseries
df = pd.DataFrame([[1, 2, 3], [100, 25, 30]]).T
df

#np conversion
X = np.array(df[0])
y = np.array(df[1])

#reshaping
X = X.reshape(3, 1)
y = y.reshape(3, 1)

from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np

regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2, degree=3))
regr.fit(X, y)

regr.predict([[4]])
```
