## synchronous
Requires the threading library

```python
#multi-threading

def thread_1_foo(val):
    #time.sleep(sleep_time)
    for a in range(val):
        print('_', 1, a)
    print('thread_1 finished')
  
def thread_2_foo(val):
    #time.sleep(sleep_time)
    for a in range(val):
        print('_', 2, a)
    print('thread_2 finished')

#creating thread
thread_1 = threading.Thread(target=thread_1_foo, args=(200,))
thread_2 = threading.Thread(target=thread_2_foo, args=(100,))

#join waits for the thread to die
#any code after join has to wait for the join process to end

#in this case we first run thread1, we wait for the completion, then thread2, completion, then regular code (print function Done!)
thread_1.start()
thread_1.join()

thread_2.start()
thread_2.join()

print("Done!")
```

```python
#single-thread

def thread_1_foo(val):
    #time.sleep(sleep_time)
    for a in range(val):
        print(a)
    print('thread_1 finished')

#creating thread
thread_1 = threading.Thread(target=thread_1_foo, args=(100,))

thread_1.start()

#join waits for the thread to die
#any code after join has to wait for the join process to end
thread_1.join()

print("Done!")
```
